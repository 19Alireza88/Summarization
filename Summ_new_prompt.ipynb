```python
import torch
from transformers import pipeline
```

```python
model_id = "meta-llama/Llama-3.2-1B"
pipe = pipeline(
    "text-generation",
    model=model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto",
)
```

```python
prompt = '''
One sentence summarization task.
Given a title and abstract of an article generate a one sentence summary of the work.

Title: Generative Inverse Design of Metamaterials with Functional Responses by Interpretable Learning
Abstract: Metamaterials with functional responses can exhibit varying properties under different conditions
(e.g., wave-based responses or deformation-induced property variation). This work addresses the rapid inverse design of
such metamaterials to meet target qualitative functional behaviors, a challenge due to its intractability and non-unique solutions.
Unlike data-intensive and non-interpretable deep-learning-based methods, we propose the Random-forest-based Interpretable Generative
Inverse Design (RIGID), a single-shot inverse design method for fast generation of metamaterial designs with on-demand functional behaviors.
RIGID leverages the interpretability of a random forest-based "design?response" forward model, eliminating the need for a more
complex "response?design" inverse model. Based on the likelihood of target satisfaction derived from the trained random forest, one can sample
a desired number of design solutions using Markov chain Monte Carlo methods. We validate RIGID on acoustic and optical metamaterial design problems,
each with fewer than 250 training samples. Compared to the genetic algorithm-based design generation approach, RIGID generates satisfactory solutions
that cover a broader range of the design space, allowing for better consideration of additional figures of merit beyond target satisfaction. This work
offers a new perspective on solving on-demand inverse design problems, showcasing the potential for incorporating interpretable machine learning into generative design under small data constraints.
One sentence summary:
'''

outputs = pipe(
    prompt,
    max_new_tokens=256,
)

print(outputs[0]["generated_text"])
```

```python
data = pd.read_csv('/content/drive/MyDrive/Project_1/new_data.csv', encoding='windows-1252')
```

```python
summ = list()
for i, j in zip(data['Title'], data['Abstract']):
  text = f'''One sentence summarization task.
Given a title and abstract of an article generate a one sentence summary of the work.
Title: {i}
Abstract: {j}
One sentence summary: '''

  outputs = pipe(
    text,
    max_new_tokens=256,)

  summ.append(outputs[0]["generated_text"])
```
